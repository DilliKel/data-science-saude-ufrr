## Explicação sobre cada coluna

* Name (Nome): Nome do paciente.
* Age (Idade): Idade do paciente em anos.
Gender (Gênero): Gênero do paciente.
Blood Type (Tipo Sanguíneo): Tipo sanguíneo do paciente.
* Medical Condition (Condição Médica): Descrição da condição médica principal do paciente.
* Date of Admission (Data de Admissão): Data em que o paciente foi admitido no hospital.
* Doctor (Médico): Nome do médico responsável pelo caso.
* Hospital (Hospital): Nome do hospital onde o paciente está sendo tratado.
* Insurance Provider (Provedor de Seguro): Nome da empresa de seguro saúde do paciente. Relevante para questões de cobertura e faturamento.
* Billing Amount (Valor da Fatura): Custo total do tratamento/internação. Importante para análises financeiras e de custos de saúde.
* Room Number (Número do Quarto): Número do quarto onde o paciente está internado. Útil para gestão hospitalar e localização do paciente.
* Admission Type (Tipo de Admissão): Categoria da admissão (ex: emergência, eletiva). Importante para análises de fluxo de pacientes e urgência.
* Discharge Date (Data de Alta): Data em que o paciente recebeu alta. Crucial para calcular o tempo de internação e eficiência do tratamento.
* Medication (Medicação): Lista de medicamentos prescritos ao paciente. Essencial para o tratamento e prevenção de interações medicamentosas.
* Test Results (Resultados de Exames): Resultados de exames médicos realizados. Fundamentais para diagnóstico, acompanhamento do tratamento e tomada de decisões médicas.

# 11. Valor Médio da Fatura, Duração Média de Estadia, Idade Média por Condições Médicas

# 12. Detalhamento da Data de Admissão

```python
data['Dia de Admissão'] = data['Data de Admissão'].dt.day
data['Mês de Admissão'] = data['Data de Admissão'].dt.month
data['Ano de Admissão'] = data['Data de Admissão'].dt.year
```

```python
data.head()
```

```python
total_revenue_by_hospital = data.groupby('Hospital')['Valor da Fatura'].sum()
top_10_hospitais = total_revenue_by_hospital.nlargest(10).index
dados_filtrados = data[data['Hospital'].isin(top_10_hospitais)]


dados_filtrados['Ano de Admissão'] = pd.to_datetime(dados_filtrados['Data de Admissão']).dt.year
faturamento_por_ano_e_hospital = dados_filtrados.groupby(['Ano de Admissão', 'Hospital'])['Valor da Fatura'].sum().reset_index()

tabela_pivo = faturamento_por_ano_e_hospital.pivot(index='Ano de Admissão', columns='Hospital', values='Valor da Fatura')
```

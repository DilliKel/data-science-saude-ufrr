Dados relacionados a **Febre Amarela em humanos e primatas não-humanos - 1994 a 2023**

Dataset a ser utilizado foi obtido em: https://opendatasus.saude.gov.br/dataset/febre-amarela-em-humanos-e-primatas-nao-humanos

Descrição dos dataset: https://s3.sa-east-1.amazonaws.com/ckan.saude.gov.br/Febre+Amarela/Dicionario+de+variaveis.pdf

```python
url = 'https://s3.sa-east-1.amazonaws.com/ckan.saude.gov.br/Febre+Amarela/fa_casoshumanos_1994-2023.csv'
df = pd.read_csv(url,encoding = "ISO-8859-1", sep=';')
```

**Conhecendo o dataset**

```python
df.head()
#Descrição dos dataset: https://s3.sa-east-1.amazonaws.com/ckan.saude.gov.br/Febre+Amarela/Dicionario+de+variaveis.pdf
```

```python
#Substituindo o NaN e valores vazios por 0 e removendo valores como string

# Replace NaN and empty values with 0
df["IDADE"] = df["IDADE"].fillna(0)
df["IDADE"] = df["IDADE"].replace('', 0)

# Remove string values (if any) and convert to numeric
df["IDADE"] = pd.to_numeric(df["IDADE"], errors='coerce').fillna(0).astype(int)
df["IDADE"].mean()
```

```python
# Ordenando o dataframe
df_ordenado = df.sort_values('UF_LPI')
df_ordenado.head()

# Conhecendo os códigos do municipios: https://www.ibge.gov.br/explica/codigos-dos-municipios.php
```

```python
"""
Seleção de colunas para manipulação dos dados
Criando um novo data frame usando df_ordenado que contenha as colunas:
UF_LPI, MUN_LPI, SEXO, IDADE, ANO_IS, OBITO e redefinindo o index do data frame
"""

novo_df = df_ordenado[['UF_LPI', 'MUN_LPI', 'SEXO', 'IDADE', 'ANO_IS', 'OBITO']].reset_index(drop=True)
novo_df.head()
```

```python
# Convertendo para lower case a coluna MUN_LPI para padronizar a comparação

novo_df['MUN_LPI'] = novo_df['MUN_LPI'].str.lower()
novo_df['MUN_LPI']
```

**Voltando para a análise**

```python
# Criar um data frame que contenha para cada municipio a latidude e longitude

# Extraindo latitude e longitude
municipalities_lat_long = municipalities[['name_muni', 'geometry', 'abbrev_state']].copy()
municipalities_lat_long['name_muni'] = municipalities_lat_long['name_muni'].str.lower()

# Extraindo latitude e longitude da coluna 'geometry'
municipalities_lat_long['latitude'] = municipalities_lat_long['geometry'].apply(
    lambda point: point.centroid.y
    )

# TODO: Extraindo longitude da coluna 'geometry'
municipalities_lat_long['longitude'] = municipalities_lat_long['geometry'].apply(
    lambda point: point.centroid.x
    )

# TODO: Removendo a coluna 'geometry'
municipalities_lat_long = municipalities_lat_long.drop(columns=['geometry'])


# Display the resulting DataFrame
municipalities_lat_long.head()
```

# Voltando a nossa análise dos dados

```python
"""
Usando municipalities_lat_long compara a coluna name_muni com a coluna MUN_LPI
em novo_df, caso verdade adicione as colunas latitude e longitude de
municipalities_lat_long em novo_df
"""
for index, row in municipalities_lat_long.iterrows():
    matching_rows = novo_df[(novo_df['MUN_LPI'] == row['name_muni']) & (novo_df['UF_LPI'] == row['abbrev_state'])]

    if not matching_rows.empty:
        novo_df.loc[matching_rows.index, 'latitude'] = row['latitude']
        # TODO: fazer a inserção de longitude
        novo_df.loc[matching_rows.index, 'longitude'] = row['longitude']
novo_df.head()
```

```python
import plotly.express as px
```

```python
"""
Plotar os casos de febre amarela no mapa do brasil com localidades
"""
# Criando o mapa
fig = px.scatter_geo(novo_df,
                     lat='latitude',
                     lon='longitude',
                     hover_name='MUN_LPI',
                     title='Casos de Febre no Brasil',
                     color_discrete_sequence=['red']
                     )

# Atualização a layer do mapa
fig.update_layout(geo=dict(
        showland=True,
        landcolor="rgb(212, 212, 212)",
        subunitcolor="rgb(255, 255, 255)",
        countrycolor="rgb(255, 255, 255)",
        showlakes=True,
        lakecolor="rgb(255, 255, 255)",
        showsubunits=True,
        showcountries=True,
        resolution=50,
        scope='south america',  # Focus on South America
        projection=dict(
            type='mercator',  # Use Mercator projection
            scale=2.5  # Adjust scale as needed
        )
    ),
    margin={"r":0,"t":30,"l":0,"b":0}
)


fig.show()
```

# Focando no estado de Roraima

```python
# TODO: Filtra o DataFrame para incluir apenas linhas onde 'UF_LPI' é igual a 'RR'


# Exibe o novo DataFrame
novo_df_rr = novo_df[novo_df['UF_LPI'] == 'RR']
novo_df_rr
```

```python
# Criando uma mapa usando novo_df_rr contendo um ponto em vermelho para cada
# coluna latitude e longitude

"""
plotar o brasil com localidades da doença
"""

import plotly.express as px

# Create the map
fig = px.scatter_geo(novo_df_rr,
                     lat='latitude',
                     lon='longitude',
                     hover_name='MUN_LPI',  # Display municipality name on hover
                     title='Casos de Febre Amarela em Roraima',
                     color_discrete_sequence=['red'])  # Set marker color to red

# Update map layout for better visualization
fig.update_layout(geo=dict(
        showland=True,
        landcolor="rgb(212, 212, 212)",
        subunitcolor="rgb(255, 255, 255)",
        countrycolor="rgb(255, 255, 255)",
        showlakes=True,
        lakecolor="rgb(255, 255, 255)",
        showsubunits=True,
        showcountries=True,
        resolution=50,
        scope='south america',  # Focus on South America
        center=dict(lat=novo_df_rr['latitude'].mean(), lon=novo_df_rr['longitude'].mean())  # Center on Roraima
    ),
    margin={"r":0,"t":30,"l":0,"b":0}
)

fig.show()
```

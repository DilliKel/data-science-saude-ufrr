### O que são dados categóricos e dados numéricos?

##### Para responder, crie uma caixa de texto abaixo e insira sua resposta nela.

---
**Dados categóricos** são variáveis que representam categorias ou grupos distintos e não possuem uma ordem específica. Eles podem ser expressos em formato de texto, como nomes, cores, tipos de produtos, etc. Exemplos incluem gênero (`masculino, feminino`), tipo de animal (`cachorro, gato, pássaro`) e estado civil (`solteiro, casado, divorciado`).

**Dados numéricos**, por outro lado, são variáveis quantitativas que representam valores numéricos. Esses valores podem ser contínuos ou discretos. Os dados numéricos podem ser manipulados matematicamente e incluem medidas como altura, peso, temperatura, idade, entre outros.

*Em resumo, enquanto dados categóricos representam diferentes categorias ou grupos, dados numéricos representam valores quantitativos.*

## Exercício 4

#### - Crie um novo DataFrame apenas com a coluna dos países e as colunas que não possuem dados faltantes e imprima as 5 primeiras entradas.

#### - A partir do DataFrame criado, busque os dados reportados pelo Brasil e armazene em um outro DataFrame chamado "casos_Brasil";

#### - A partir do DataFrame "casos_Brasil", exiba a média de casos acumulados e mortes acumuladas;

##### * dica de Resposta: utilize uma célula de código para cada item da resposta.

```python
# Criar um novo DataFrame apenas com as colunas dos países e as colunas que não possuem dados faltantes
novo_df = dados[['Country', 'Date_reported', 'New_cases', 'Cumulative_cases', 'New_deaths', 'Cumulative_deaths']].dropna()

# Imprimir as 5 primeiras entradas do novo DataFrame
print(novo_df.head())
```

```python
# Criar um DataFrame apenas com os dados reportados pelo Brasil
casos_Brasil = novo_df[novo_df['Country'] == 'Brazil']

# Imprimir as 5 primeiras entradas do DataFrame "casos_Brasil"
print(casos_Brasil.head())
```

```python
# Calcular a média de casos acumulados e mortes acumuladas no Brasil
media_casos = casos_Brasil['Cumulative_cases'].mean()
media_mortes = casos_Brasil['Cumulative_deaths'].mean()

# Imprimir a média de casos acumulados e mortes acumuladas no Brasil
print("Média de casos acumulados no Brasil:", media_casos)
print("Média de mortes acumuladas no Brasil:", media_mortes)
```

## Exercício 5

#### - Crie um novo DataFrame apenas com o país que contém o maior número de mortes acumuladas em um único dia;

#### - A partir do DataFrame criado, crie um outro (sugestão de nome "pais_maior_numero_mortes") apenas com as colunas que não possuem dados faltantes e imprima as 5 primeiras entradas;

#### - A partir do DataFrame "pais_maior_numero_mortes", exiba a média de casos acumulados e mortes acumuladas;

#### - Agora plot um gráfico de barras que exiba os casos acumulados por países;

#### - Por fim, plot um histograma que exiba as mortes acumuladas e sua respectiva frequência;

##### * dica de Resposta: utilize uma célula de código para cada item da resposta.

```python
# Encontrar o país com o maior número de mortes acumuladas em um único dia
pais_maior_mortes = dados.loc[dados['Cumulative_deaths'].idxmax()]['Country'] # Use 'dados' ao invés de 'df'

# Criar um novo DataFrame apenas com os dados do país com o maior número de mortes acumuladas
df_pais_maior_mortes = dados[dados['Country'] == pais_maior_mortes] # Use 'dados' ao invés de 'df'

# Imprimir o nome do país com o maior número de mortes acumuladas
print("País com o maior número de mortes acumuladas em um único dia:", pais_maior_mortes)
```

```python
# Criar um novo DataFrame apenas com as colunas que não possuem dados faltantes
pais_maior_numero_mortes = df_pais_maior_mortes[['Country', 'Date_reported', 'New_cases', 'Cumulative_cases', 'New_deaths', 'Cumulative_deaths']].dropna()

# Imprimir as 5 primeiras entradas do novo DataFrame
print(pais_maior_numero_mortes.head())
```

```python
# Calcular a média de casos acumulados e mortes acumuladas no país com o maior número de mortes acumuladas
media_casos_pais = pais_maior_numero_mortes['Cumulative_cases'].mean()
media_mortes_pais = pais_maior_numero_mortes['Cumulative_deaths'].mean()

# Imprimir a média de casos acumulados e mortes acumuladas no país com o maior número de mortes acumuladas
print("Média de casos acumulados no país com o maior número de mortes acumuladas:", media_casos_pais)
print("Média de mortes acumuladas no país com o maior número de mortes acumuladas:", media_mortes_pais)
```

```python
import matplotlib.pyplot as plt

# Filtrar os dados para excluir os casos onde as mortes acumuladas são iguais a zero
dados_filtrados = dados[dados['Cumulative_deaths'] != 0]

# Calcular os casos acumulados por país
casos_acumulados_por_pais = dados_filtrados.groupby('Country')['Cumulative_cases'].max()

# Plotar o gráfico de barras
plt.figure(figsize=(55, 6))
casos_acumulados_por_pais.plot(kind='bar', color='skyblue')
plt.title('Casos acumulados por país (excluindo casos com zero mortes acumuladas)')
plt.xlabel('País')
plt.ylabel('Casos acumulados')
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.grid(axis='y')  # Adiciona linhas de grade apenas no eixo y
plt.tight_layout()  # Ajusta automaticamente o layout para evitar que os rótulos fiquem sobrepostos
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotar o histograma
plt.figure(figsize=(10, 6))
plt.hist(dados['Cumulative_deaths'], bins=30, color='skyblue', edgecolor='black') # Use 'dados' ao invés de 'df'
plt.title('Histograma de mortes acumuladas')
plt.xlabel('Mortes acumuladas')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

```python
# Importando a biblioteca pandas
import pandas as pd

# Carregando os datasets
confirmados = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
mortes = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recuperados = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')

# Verificando as 5 primeiras linhas de cada DataFrame
print("Casos Confirmados:")
print(confirmados.head())

print("\nMortes:")
print(mortes.head())

print("\nRecuperados:")
print(recuperados.head())
```

```python
# Calculando o total de casos confirmados por país
confirmados_por_pais = confirmados.groupby('Country/Region').sum().iloc[:, 3:]
confirmados_por_pais['Total_Confirmados'] = confirmados_por_pais.sum(axis=1)
confirmados_por_pais = confirmados_por_pais[['Total_Confirmados']]

# Exibindo as 5 primeiras linhas do DataFrame de casos confirmados por país
print("Confirmados por País:")
print(confirmados_por_pais.head())
```

```python
# Calculando o total de mortes por país
mortes_por_pais = mortes.groupby('Country/Region').sum().iloc[:, 3:]
mortes_por_pais['Total_Mortes'] = mortes_por_pais.sum(axis=1)
mortes_por_pais = mortes_por_pais[['Total_Mortes']]

# Exibindo as 5 primeiras linhas do DataFrame de mortes por país
print("Mortes por País:")
print(mortes_por_pais.head())
```

```python
# Calculando o total de recuperados por país
recuperados_por_pais = recuperados.groupby('Country/Region').sum().iloc[:, 3:]
recuperados_por_pais['Total_Recuperados'] = recuperados_por_pais.sum(axis=1)
recuperados_por_pais = recuperados_por_pais[['Total_Recuperados']]

# Exibindo as 5 primeiras linhas do DataFrame de recuperados por país
print("Recuperados por País:")
print(recuperados_por_pais.head())
```

```python
# Criando uma tabela unindo as bases
nova_DB = [confirmados_por_pais, mortes_por_pais, recuperados_por_pais]
base_combinada = pd.concat(nova_DB, axis=1)

# Mudando o nome das colunas
base_combinada.columns = ["Confirmados", "Mortes", "Recuperados"]
print("\nBase Combinada:")
print(base_combinada.head())
```

```python
# Verificando os últimos 5 itens do DataFrame combinado
print("\nÚltimos itens da Base Combinada:")
print(base_combinada.tail())
```

### Exercício 3
- Qual o país com maior número de casos confirmados?
- Qual o país com maior número de mortes?

```python
# Identificar o País com Maior Número de Casos Confirmados e Mortes

# País com maior número de casos confirmados
pais_maior_confirmados = base_combinada['Confirmados'].idxmax()
maior_confirmados = base_combinada['Confirmados'].max()
print(f"O país com maior número de casos confirmados é {pais_maior_confirmados} com {maior_confirmados} casos.")

# País com maior número de mortes
pais_maior_mortes = base_combinada['Mortes'].idxmax()
maior_mortes = base_combinada['Mortes'].max()
print(f"O país com maior número de mortes é {pais_maior_mortes} com {maior_mortes} mortes.")
```

### Exercício 5

- Calcule a taxa de mortalidade e encontre os países com as maiores e menores taxas.
- Calcule a taxa de letalidade e encontre os países com as maiores e menores taxas.

```python
# Criando um DataFrame de exemplo para populações
# Em um cenário real, você deveria carregar isso de uma fonte confiável
populacao_data = {
    'Country/Region': ['US', 'Brazil', 'Russia', 'India', 'UK'],
    'Populacao': [331002651, 212559417, 145934462, 1380004385, 67886011]
}
populacao = pd.DataFrame(populacao_data)
populacao.set_index('Country/Region', inplace=True)

# Adicionando a população ao DataFrame combinado
base_combinada = base_combinada.join(populacao)

# Calculando a taxa de mortalidade (por 1000 habitantes)
base_combinada['Taxa_Mortalidade'] = (base_combinada['Mortes'] / base_combinada['Populacao']) * 1000

# Exibindo a taxa de mortalidade
print("Taxa de Mortalidade por País:")
print(base_combinada[['Mortes', 'Populacao', 'Taxa_Mortalidade']].dropna().head())
```

```python
# Calculando a taxa de letalidade (por 100 casos confirmados)
base_combinada['Taxa_Letalidade'] = (base_combinada['Mortes'] / base_combinada['Confirmados']) * 100

# Exibindo a taxa de letalidade
print("Taxa de Letalidade por País:")
print(base_combinada[['Mortes', 'Confirmados', 'Taxa_Letalidade']].dropna().head())
```

```python
# Países com as maiores taxas de mortalidade
maiores_taxa_mortalidade = base_combinada.nlargest(5, 'Taxa_Mortalidade')
print("Países com as Maiores Taxas de Mortalidade:")
print(maiores_taxa_mortalidade[['Taxa_Mortalidade']])

# Países com as menores taxas de mortalidade
menores_taxa_mortalidade = base_combinada.nsmallest(5, 'Taxa_Mortalidade')
print("\nPaíses com as Menores Taxas de Mortalidade:")
print(menores_taxa_mortalidade[['Taxa_Mortalidade']])

# Países com as maiores taxas de letalidade
maiores_taxa_letalidade = base_combinada.nlargest(5, 'Taxa_Letalidade')
print("\nPaíses com as Maiores Taxas de Letalidade:")
print(maiores_taxa_letalidade[['Taxa_Letalidade']])

# Países com as menores taxas de letalidade
menores_taxa_letalidade = base_combinada.nsmallest(5, 'Taxa_Letalidade')
print("\nPaíses com as Menores Taxas de Letalidade:")
print(menores_taxa_letalidade[['Taxa_Letalidade']])
```

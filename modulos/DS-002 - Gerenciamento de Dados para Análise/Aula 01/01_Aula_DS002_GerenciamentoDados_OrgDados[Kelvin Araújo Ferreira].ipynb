##### **07. Descubra os casos de NaN e NULL na base de dados.**

> Podem ocorrer casos de valores que não são números e de valores nulos. A depender de casa situação, pode ser necessário remover ambos. É o que faremos aqui.

Vamos começar importando os dados e criando uma cópia do DataFrame original.

```python
import pandas as pd

# Carregar a base de dados original
dados = pd.read_csv('/content/drive/MyDrive/Ciência da Computação/Data Science - Projeto de Extensão/DS-002 - Gerenciamento de Dados para Análise/Aula 01/dengue05.csv')

# Criar uma cópia do DataFrame original
df_dengue = dados.copy()
```

Aqui, df_dengue será uma cópia exata do DataFrame dados.



```python
# Renomear as colunas do DataFrame df_dengue
df_dengue.rename(columns={
    'Year': 'Ano',
    'State': 'Estado',
    'Population': 'População',
    'Region': 'Região',
    'Min_Tmp': 'Temp_Minima',
    'Max_Tmp': 'Temp_Maxima',
    'Change_Tmp': 'Variacao_Temp',
    'Mean_Tmp': 'Temp_Media',
    'Percipitation': 'Precipitacao',
    'Cases': 'Casos'
}, inplace=True)
```

```python
# Exibir as primeiras linhas do DataFrame df_dengue
df_dengue.head()
```

```python
# Verificar informações do DataFrame
df_dengue.info()
```

```python
# Remover linhas com valores ausentes, se necessário
df_dengue.dropna(inplace=True)

# Calcular a mediana dos casos de dengue
mediana_casos = df_dengue['Casos'].median()

# Definir função para identificar se há muitos casos
def ident_se_muitos_casos(casos):
    if casos > mediana_casos:
        return 1
    else:
        return 0

# Criar a coluna 'SeveralCases' com base na função definida
df_dengue['SeveralCases'] = df_dengue['Casos'].apply(ident_se_muitos_casos)
```

```python
# Exibir a nova coluna 'SeveralCases'
print("\nNova coluna 'SeveralCases':")
print(df_dengue['SeveralCases'])
```

> **Para cálculo de outliers, podemos usar a seguinte estratégia:**

```
# Exemplo de DataFrame
dados = pd.DataFrame({'valores': [10, 15, 20, 25, 30, 100, 105, 110, 115]})

# Cálculo dos quartis
q1 = dados['valores'].quantile(0.25)
q3 = dados['valores'].quantile(0.75)

# Cálculo do IQR
iqr = q3 - q1

# Cálculo dos limites superior e inferior
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

# Identificação dos outliers
outliers = dados[(dados['valores'] < limite_inferior) | (dados['valores'] > limite_superior)]
```

##### **02. Calcule a correlação entre variáveis importantes do Dataset e que poderão compor o seu modelo.**

💡**Um pouco de teoria sobre correlações**
- **Variável** é qualquer fenômeno cujo valor possa mudar em, pelo menos, dois níveis, estados ou graus. Exemplos de variáveis: temperatura, velocidade, altura, peso, idade, gravidade da doença etc.
- A **correlação** é uma medida de associação entre duas variáveis. Essa associação pode variar de 0 a 1. Quando o valor é muito próximo de 0, interpretamos que a associação é muita baixa ou inexistente. Por outro lado, quanto mais próximo de 1, interpretamos que existe uma associação entre as variáveis.    
- Uma associação entre a Variável 1 e a Variável 2 pode ser ***diretamente proporcional*** ou ***inversamente proporcional***.
- A associação diretamente proporcional indica que quanto mais a Variável 1 tem seu valor aumentado, mais isso também é observado na Variável 2 e vice-versa.
- A associação inversamente proporcional indica que quanto mais a Variável 1 aumenta, mais a Variável 2 reduz o seu valor e vice-versa.

**A imagem a seguir explicita esses dois tipos de relações:**
![Tipos de Correlação](https://edtisensei.zendesk.com/hc/article_attachments/360039155971/mceclip1.png)

````python
df_dengue[['Change_Tmp',
           'Max_Tmp',
           'Min_Tmp',
           'Mean_Tmp',
           'Percipitation',
           'Cases',
           '%CasesInPop']].corr(method='pearson')
````

🤓**Observações importantes:**

- Muitas vezes queremos prever uma variável ou classificar um conjunto de dados usando uma variável. Usamos Aprendizado de Máquina / Inteligência Artificial para fazer isso.
- Quando você for selecionar variáveis para um modelo de IA, procure evitar manter duas variáveis que estejam fortemente correlacionadas entre si. Essa redundância prejudica o modelo. Tal situação está associada ao conceito de multicolinearidade. Existem técnicas estatísticas para avaliar a existência desse fenômeno. Não obstante, um exame da matriz de correlação pode fornecer indícios da existência desse problema.

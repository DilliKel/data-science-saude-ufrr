```python
# @title Reiniciando o notebook
import pandas as pd

dados = pd.read_csv('/content/drive/MyDrive/Ciência da Computação/Data Science - Projeto de Extensão/DS-002 - Gerenciamento de Dados para Análise/Aula 02/dengue.csv')

df_dengue = pd.DataFrame(dados, columns = ['Year', 'State', 'Population', 'Region', 'Min_Tmp',
                                               'Max_Tmp', 'Change_Tmp', 'Mean_Tmp', 'Percipitation','Cases'])
df_dengue = df_dengue.dropna()

mediana = df_dengue['Cases'].median()
def ident_se_muitos_casos(casos):
  if casos > mediana:
    return 1
  else:
    return 0

df_dengue['SeveralCases'] = df_dengue['Cases'].apply(ident_se_muitos_casos)
```

```python
# Criar a coluna 'FromMainStates'
main_states = ['Distrito Federal', 'Minas Gerais', 'Paraná']
df_dengue['FromMainStates'] = df_dengue['State'].apply(lambda x: 1 if x in main_states else 0)
```

##### **29. Transforme os dados da coluna FromMainStates em variáveis dummy.**

> Vários modelos de inteligência artificial, como a Árvore de Decisão, não irão processar valores de texto como os contidos na coluna Region. Felizmente, podemos criar uma coluna para cada região (N, NE, S, SE e CO), que, por exemplo, na coluna da região Norte, conterá o número 1 quando a linha do banco for referente ao Norte e 0, caso contrário.

```python
# Transforme os dados da coluna FromMainStates em variáveis dummy.

import pandas as pd
df_dummies = pd.get_dummies(df_dengue_sem[['FromMainStates']])
df_dummies.head()
```

##### **30. Mude o nome das novas colunas de features criadas, reordene as colunas para as novas ficarem no início do banco e verifique como ficou o DataFrame `features_encoded` após esse conjunto de mudanças.**

```python
import pandas as pd
# Mude o nome das novas colunas de features criadas
df_dummies = df_dummies.rename(columns={'FromMainStates_0': 'FromMainStates_0_dummy', 'FromMainStates_1': 'FromMainStates_1_dummy'})

# Reordene as colunas para as novas ficarem no início do banco
df_dummies = pd.concat([df_dummies, df_features], axis=1)

# Verifique como ficou o DataFrame features_encoded após esse conjunto de mudanças
df_dummies.head()
```

criando uma colunade classificação do imc de cada indivíduo

```python
def classificar_imc(imc):
    if imc < 16.0:
        return 1  # Baixo peso severo
    elif 16.0 <= imc < 16.9:
        return 2  # Baixo peso moderado
    elif 17.0 <= imc < 18.4:
        return 3  # Baixo peso leve
    elif 18.5 <= imc < 24.9:
        return 4  # Peso normal
    elif 25.0 <= imc < 29.9:
        return 5  # Sobrepeso
    elif 30.0 <= imc < 34.9:
        return 6  # Obesidade grau I
    elif 35.0 <= imc < 39.9:
        return 7  # Obesidade grau II (severa)
    else:
        return 8  # Obesidade grau III (mórbida)
```

```python
df["classificacao_IMC"] = df["imc"].apply(classificar_imc)

df.head()
```

```python
count_class_imc = df['classificacao_IMC'].value_counts()

print(count_class_imc)
```
